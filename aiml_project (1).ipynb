```python
import numpy as np
import pandas as pd

class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0

        # Gradient descent
        for _ in range(self.num_iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            y_predicted = self.sigmoid(linear_model)

            # Compute gradients
            dw = (1 / num_samples) * np.dot(X.T, (y_predicted - y))
            db = (1 / num_samples) * np.sum(y_predicted - y)

            # Update parameters
            self.weights -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        y_predicted = self.sigmoid(linear_model)
        y_predicted_cls = [1 if i > 0.5 else 0 for i in y_predicted]
        return np.array(y_predicted_cls)

    def evaluate(self, X, y):
        y_predicted = self.predict(X)
        accuracy = np.mean(y_predicted == y)

        precision = self.calculate_precision(y, y_predicted)
        recall = self.calculate_recall(y, y_predicted)
        f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        return accuracy, precision, recall, f1_score, y_predicted

    def calculate_precision(self, y_true, y_pred):
        true_positives = np.sum((y_true == 1) & (y_pred == 1))
        predicted_positives = np.sum(y_pred == 1)

        if predicted_positives == 0:
            return 0
        else:
            return true_positives / predicted_positives

    def calculate_recall(self, y_true, y_pred):
        true_positives = np.sum((y_true == 1) & (y_pred == 1))
        actual_positives = np.sum(y_true == 1)

        if actual_positives == 0:
            return 0
        else:
            return true_positives / actual_positives

def train_and_evaluate_model(X_train, y_train, X_test, y_test, learning_rate=0.01, num_iterations=1000):
    # Training
    model = LogisticRegression(learning_rate, num_iterations)
    model.fit(X_train, y_train)

    # Evaluation
    accuracy, precision, recall, f1_score, _ = model.evaluate(X_test, y_test)
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1_score)

    return model

def get_user_input_and_predict(model, feature_names):
    print("\nPlease enter the features for which you want predictions:")
    user_input = []
    for i, feature_name in enumerate(feature_names):
        user_input.append(float(input(f"Enter value for feature '{feature_name}': ")))

    # Make predictions based on user input
    user_input = np.array(user_input).reshape(1, -1)
    user_prediction = model.predict(user_input)
    print("Predicted outcome based on user input:", user_prediction)

    return user_prediction

# Load dataset
try:
    data = pd.read_csv('/content/data (2).csv')
except FileNotFoundError:
    print("Error: File not found. Please provide the correct path to the CSV file.")
    exit()

# Check if the target column exists
target_column = 'verification.result'
if target_column not in data.columns:
    print(f"Error: Target column '{target_column}' not found in the dataset.")
    exit()

# Preprocessing
X = data.drop(target_column, axis=1).values
y = data[target_column].values

# Feature scaling
X = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Split dataset into training (80%) and testing (20%) sets
split_ratio = 0.8
split_index = int(split_ratio * len(X))

X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

# Train and evaluate the model
model = train_and_evaluate_model(X_train, y_train, X_test, y_test)

# Get input from the user and make predictions
user_prediction = get_user_input_and_predict(model, data.columns[:-1])  # Exclude the target column for input

# Calculate precision, recall, and F1-score for the test set
_, precision_test, recall_test, f1_score_test, _ = model.evaluate(X_test, y_test)
print(f"Precision (Test Set): {precision_test:.4f}")
print(f"Recall (Test Set): {recall_test:.4f}")
print(f"F1 Score (Test Set): {f1_score_test:.4f}")
```
